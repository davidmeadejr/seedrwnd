```python
import pandas as pd
from ntscraper import Nitter
```

```python
scraper = Nitter()
```

```python
def get_tweets():
    terms = ['Startup idea:']
    results = scraper.get_tweets(terms, mode='term', number=834, since='2018-12-20', until='2023-12-20')
    
    # Access the 'tweets' key in the results dictionary
    tweet_data = results['tweets']

    final_tweets = []
    # Process each tweet in the tweet_data list
    for tweet in tweet_data:
        data = [tweet['link'], tweet['text'], tweet['date'], tweet['stats']['likes'], tweet['stats']['comments']]
        final_tweets.append(data)

    df = pd.DataFrame(final_tweets, columns=['link', 'text', 'date', 'Likes', 'Comments'])
    return df

tweets_df = get_tweets()
```

```python
tweets_df
```

```python
tweets_df.to_csv('startup_idea_tweets.csv', index=False)
```
